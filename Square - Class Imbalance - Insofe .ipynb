```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,precision_score, f1_score, recall_score, accuracy_score,classification_report,precision_recall_fscore_support
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
def all_model_report(model_names,reports):
    report_data = []
    for i,report in enumerate(reports):
        lines = report.split('\n')
        for line in lines[2:5]:
            row = {}
            row_data = line.split('      ')
            row['Model'] = model_names[i]
            row['class'] ="Class"+row_data[1]
            row['precision'] = float(row_data[2])
            row['recall'] = float(row_data[3])
            row['f1_score'] = float(row_data[4])
            row['support'] = float(row_data[5])
            report_data.append(row)
    return report_data
def split(data_set):
    X = data_set.drop('Target',axis = 1)
    y = data_set['Target']
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify = y,random_state=0)
    models = [['Logistic Regression : \n\n',LogisticRegression(random_state = 0, solver = 'lbfgs',multi_class = 'multinomial')],
              ['Gaussian Naive Bayes : \n\n',GaussianNB()],
              ['Decision_Tree_Classifier : \n\n,',DecisionTreeClassifier(random_state = 0)],
              ['Random_Forest : \n\n',RandomForestClassifier(random_state = 0)],
              ['Knn : \n\n',KNeighborsClassifier()],
              ['Xg_boost : \n\n',XGBClassifier(random_state = 0)],
              ['Ada_Boost : \n\n',AdaBoostClassifier(random_state = 0)],
              ['Gradient Boosting Classification : \n\n', GradientBoostingClassifier(random_state = 0)],
              ['Bagging Classifier : \n\n',BaggingClassifier(random_state = 0)]]
    report = []
    for name,model in models:
        model = model
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        class_report = classification_report(y_test,y_pred)
        report.append(class_report)       
    return report
```
